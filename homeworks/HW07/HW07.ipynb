```python
# -*- coding: utf-8 -*-
"""
–î–æ–º–∞—à–Ω—è—è —Ä–∞–±–æ—Ç–∞ ‚Ññ7: –ö–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏—è
–î–∞—Ç–∞—Å–µ—Ç—ã: S07-hw-dataset-01.csv, S07-hw-dataset-02.csv, S07-hw-dataset-03.csv
–°—Ç—É–¥–µ–Ω—Ç: –†–µ–≥–∏–Ω–∞
"""

# 1. –ò–ú–ü–û–†–¢ –ë–ò–ë–õ–ò–û–¢–ï–ö
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
import os
import warnings
warnings.filterwarnings('ignore')

from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.metrics import (silhouette_score, davies_bouldin_score, 
                           calinski_harabasz_score, adjusted_rand_score)

# –ù–∞—Å—Ç—Ä–æ–π–∫–∞ —Å—Ç–∏–ª–µ–π –≥—Ä–∞—Ñ–∏–∫–æ–≤
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 12

print("‚úÖ –ë–∏–±–ª–∏–æ—Ç–µ–∫–∏ –∑–∞–≥—Ä—É–∂–µ–Ω—ã")

# 2. –°–û–ó–î–ê–ù–ò–ï –°–¢–†–£–ö–¢–£–†–´ –ü–ê–ü–û–ö
def create_project_structure():
    """–°–æ–∑–¥–∞—ë—Ç –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–µ –ø–∞–ø–∫–∏ –¥–ª—è –∞—Ä—Ç–µ—Ñ–∞–∫—Ç–æ–≤"""
    folders = [
        'artifacts/figures',
        'artifacts/labels'
    ]
    
    for folder in folders:
        os.makedirs(folder, exist_ok=True)
        print(f"‚úì –ü–∞–ø–∫–∞ —Å–æ–∑–¥–∞–Ω–∞/–ø—Ä–æ–≤–µ—Ä–µ–Ω–∞: {folder}")
    
    print("\n‚úÖ –°—Ç—Ä—É–∫—Ç—É—Ä–∞ –ø–∞–ø–æ–∫ –≥–æ—Ç–æ–≤–∞")

create_project_structure()

# 3. –§–£–ù–ö–¶–ò–Ø –î–õ–Ø –ê–ù–ê–õ–ò–ó–ê –û–î–ù–û–ì–û –î–ê–¢–ê–°–ï–¢–ê
def analyze_dataset(dataset_name, dataset_path, dataset_num):
    """
    –ê–Ω–∞–ª–∏–∑–∏—Ä—É–µ—Ç –æ–¥–∏–Ω –¥–∞—Ç–∞—Å–µ—Ç –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã
    
    Parameters:
    -----------
    dataset_name : str
        –ù–∞–∑–≤–∞–Ω–∏–µ –¥–∞—Ç–∞—Å–µ—Ç–∞
    dataset_path : str
        –ü—É—Ç—å –∫ —Ñ–∞–π–ª—É –¥–∞—Ç–∞—Å–µ—Ç–∞
    dataset_num : int
        –ù–æ–º–µ—Ä –¥–∞—Ç–∞—Å–µ—Ç–∞ (1, 2, 3)
    
    Returns:
    --------
    dict : –°–ª–æ–≤–∞—Ä—å —Å —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º–∏ –∞–Ω–∞–ª–∏–∑–∞
    """
    
    print(f"\n{'='*60}")
    print(f"–ê–ù–ê–õ–ò–ó –î–ê–¢–ê–°–ï–¢–ê {dataset_num}: {dataset_name}")
    print(f"{'='*60}")
    
    # –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞—Ç–∞—Å–µ—Ç–∞
    try:
        df = pd.read_csv(dataset_path)
        print(f"‚úÖ –î–∞—Ç–∞—Å–µ—Ç –∑–∞–≥—Ä—É–∂–µ–Ω: {df.shape[0]} —Å—Ç—Ä–æ–∫, {df.shape[1]} —Å—Ç–æ–ª–±—Ü–æ–≤")
    except Exception as e:
        print(f"‚ùå –û—à–∏–±–∫–∞ –ø—Ä–∏ –∑–∞–≥—Ä—É–∑–∫–µ {dataset_name}: {e}")
        return None
    
    # –ü–µ—Ä–≤–∏—á–Ω—ã–π –∞–Ω–∞–ª–∏–∑
    print(f"\nüìä –ü–µ—Ä–≤—ã–µ 5 —Å—Ç—Ä–æ–∫:")
    display(df.head())
    
    print(f"\nüìã –ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –¥–∞–Ω–Ω—ã—Ö:")
    print(df.info())
    
    # –ü—Ä–æ–≤–µ—Ä–∫–∞ –ø—Ä–æ–ø—É—Å–∫–æ–≤
    missing = df.isnull().sum()
    missing_percent = (missing / len(df)) * 100
    print(f"\nüîç –ü—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è:")
    missing_df = pd.DataFrame({
        '–ü—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π': missing,
        '–î–æ–ª—è, %': missing_percent.round(2)
    })
    display(missing_df[missing_df['–ü—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π'] > 0])
    
    # –°–æ—Ö—Ä–∞–Ω—è–µ–º sample_id –æ—Ç–¥–µ–ª—å–Ω–æ
    if 'sample_id' in df.columns:
        sample_ids = df['sample_id']
        X = df.drop('sample_id', axis=1)
        print("‚úì –°—Ç–æ–ª–±–µ—Ü 'sample_id' —Å–æ—Ö—Ä–∞–Ω—ë–Ω –æ—Ç–¥–µ–ª—å–Ω–æ")
    else:
        sample_ids = pd.Series(range(len(df)))
        X = df.copy()
        print("‚ö†Ô∏è –°—Ç–æ–ª–±–µ—Ü 'sample_id' –Ω–µ –Ω–∞–π–¥–µ–Ω, —Å–æ–∑–¥–∞–Ω –∏—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω—ã–π")
    
    # –û–ø—Ä–µ–¥–µ–ª—è–µ–º —Ç–∏–ø—ã –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    numeric_features = X.select_dtypes(include=[np.number]).columns.tolist()
    categorical_features = X.select_dtypes(include=['object']).columns.tolist()
    
    print(f"\nüìå –ß–∏—Å–ª–æ–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ ({len(numeric_features)}):")
    print(numeric_features[:10])  # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º –ø–µ—Ä–≤—ã–µ 10
    
    print(f"\nüìå –ö–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ ({len(categorical_features)}):")
    print(categorical_features)
    
    # –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —á–∏—Å–ª–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    if len(numeric_features) > 0:
        n_features_to_plot = min(6, len(numeric_features))
        fig, axes = plt.subplots(2, 3, figsize=(15, 10))
        axes = axes.flatten()
        
        for i, col in enumerate(numeric_features[:n_features_to_plot]):
            axes[i].hist(X[col].dropna(), bins=30, edgecolor='black', alpha=0.7)
            axes[i].set_title(f'{dataset_name}: {col}')
            axes[i].set_xlabel('–ó–Ω–∞—á–µ–Ω–∏–µ')
            axes[i].set_ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
        
        # –°–∫—Ä—ã–≤–∞–µ–º –ø—É—Å—Ç—ã–µ subplots
        for i in range(n_features_to_plot, len(axes)):
            axes[i].set_visible(False)
        
        plt.tight_layout()
        plt.savefig(f'artifacts/figures/{dataset_name}_distribution.png', dpi=300, bbox_inches='tight')
        plt.show()
    
    # 4. –ü–†–ï–ü–†–û–¶–ï–°–°–ò–ù–ì
    print(f"\n{'='*60}")
    print(f"–ü–†–ï–ü–†–û–¶–ï–°–°–ò–ù–ì –î–ê–¢–ê–°–ï–¢–ê {dataset_num}")
    print(f"{'='*60}")
    
    # –°–æ–∑–¥–∞—ë–º –ø–∞–π–ø–ª–∞–π–Ω –¥–ª—è –ø—Ä–µ–ø—Ä–æ—Ü–µ—Å—Å–∏–Ω–≥–∞
    numeric_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler())
    ])
    
    preprocessor = ColumnTransformer(
        transformers=[
            ('num', numeric_transformer, numeric_features)
        ],
        remainder='drop'  # –ò–≥–Ω–æ—Ä–∏—Ä—É–µ–º –Ω–µ—á–∏—Å–ª–æ–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏
    )
    
    print("‚öôÔ∏è –ü—Ä–∏–º–µ–Ω—è–µ–º –ø—Ä–µ–ø—Ä–æ—Ü–µ—Å—Å–∏–Ω–≥...")
    X_processed = preprocessor.fit_transform(X)
    print(f"‚úì –î–∞–Ω–Ω—ã–µ –æ–±—Ä–∞–±–æ—Ç–∞–Ω—ã. –†–∞–∑–º–µ—Ä –ø–æ—Å–ª–µ –ø—Ä–µ–ø—Ä–æ—Ü–µ—Å—Å–∏–Ω–≥–∞: {X_processed.shape}")
    
    # 5. KMEANS –ö–õ–ê–°–¢–ï–†–ò–ó–ê–¶–ò–Ø
    print(f"\n{'='*60}")
    print(f"KMEANS –ö–õ–ê–°–¢–ï–†–ò–ó–ê–¶–ò–Ø –î–ê–¢–ê–°–ï–¢–ê {dataset_num}")
    print(f"{'='*60}")
    
    # –ü–æ–¥–±–æ—Ä –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–≥–æ —á–∏—Å–ª–∞ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤
    k_range = range(2, 21)
    silhouette_scores = []
    db_scores = []
    ch_scores = []
    inertia_values = []
    
    print(f"üîç –ü–æ–¥–±–∏—Ä–∞–µ–º –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–µ k –≤ –¥–∏–∞–ø–∞–∑–æ–Ω–µ {min(k_range)}-{max(k_range)}...")
    
    for k in k_range:
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        labels = kmeans.fit_predict(X_processed)
        
        silhouette_scores.append(silhouette_score(X_processed, labels))
        db_scores.append(davies_bouldin_score(X_processed, labels))
        ch_scores.append(calinski_harabasz_score(X_processed, labels))
        inertia_values.append(kmeans.inertia_)
    
    # –í—ã–±–∏—Ä–∞–µ–º –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–µ k
    optimal_k_silhouette = k_range[np.argmax(silhouette_scores)]
    print(f"‚úÖ –û–ø—Ç–∏–º–∞–ª—å–Ω–æ–µ k –ø–æ Silhouette: {optimal_k_silhouette}")
    
    # –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –º–µ—Ç—Ä–∏–∫ KMeans
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    
    # 1. Silhouette Score
    axes[0, 0].plot(k_range, silhouette_scores, marker='o', linewidth=2, markersize=8)
    axes[0, 0].set_xlabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ (k)', fontsize=12)
    axes[0, 0].set_ylabel('Silhouette Score', fontsize=12)
    axes[0, 0].set_title(f'{dataset_name}: –ö–∞—á–µ—Å—Ç–≤–æ –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏ (Silhouette)', fontsize=14)
    axes[0, 0].grid(True, alpha=0.3)
    axes[0, 0].axvline(x=optimal_k_silhouette, color='r', linestyle='--', 
                       label=f'–û–ø—Ç–∏–º–∞–ª—å–Ω–æ–µ k={optimal_k_silhouette}')
    axes[0, 0].legend()
    
    # 2. Davies-Bouldin Score
    axes[0, 1].plot(k_range, db_scores, marker='s', color='orange', 
                    linewidth=2, markersize=8)
    axes[0, 1].set_xlabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ (k)', fontsize=12)
    axes[0, 1].set_ylabel('Davies-Bouldin Score', fontsize=12)
    axes[0, 1].set_title(f'{dataset_name}: –ö–æ–º–ø–∞–∫—Ç–Ω–æ—Å—Ç—å –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ (Davies-Bouldin)', fontsize=14)
    axes[0, 1].grid(True, alpha=0.3)
    optimal_k_db = k_range[np.argmin(db_scores)]
    axes[0, 1].axvline(x=optimal_k_db, color='r', linestyle='--', 
                       label=f'–û–ø—Ç–∏–º–∞–ª—å–Ω–æ–µ k={optimal_k_db}')
    axes[0, 1].legend()
    
    # 3. Calinski-Harabasz Score
    axes[1, 0].plot(k_range, ch_scores, marker='^', color='green', 
                    linewidth=2, markersize=8)
    axes[1, 0].set_xlabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ (k)', fontsize=12)
    axes[1, 0].set_ylabel('Calinski-Harabasz Score', fontsize=12)
    axes[1, 0].set_title(f'{dataset_name}: –û—Ç–Ω–æ—à–µ–Ω–∏–µ –¥–∏—Å–ø–µ—Ä—Å–∏–π (Calinski-Harabasz)', fontsize=14)
    axes[1, 0].grid(True, alpha=0.3)
    optimal_k_ch = k_range[np.argmax(ch_scores)]
    axes[1, 0].axvline(x=optimal_k_ch, color='r', linestyle='--', 
                       label=f'–û–ø—Ç–∏–º–∞–ª—å–Ω–æ–µ k={optimal_k_ch}')
    axes[1, 0].legend()
    
    # 4. Elbow Method
    axes[1, 1].plot(k_range, inertia_values, marker='D', color='purple', 
                    linewidth=2, markersize=8)
    axes[1, 1].set_xlabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ (k)', fontsize=12)
    axes[1, 1].set_ylabel('Inertia', fontsize=12)
    axes[1, 1].set_title(f'{dataset_name}: –ú–µ—Ç–æ–¥ –ª–æ–∫—Ç—è (Elbow Method)', fontsize=14)
    axes[1, 1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig(f'artifacts/figures/{dataset_name}_kmeans_metrics.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # –û–±—É—á–∞–µ–º KMeans —Å –≤—ã–±—Ä–∞–Ω–Ω—ã–º k
    chosen_k = optimal_k_silhouette
    kmeans_best = KMeans(n_clusters=chosen_k, random_state=42, n_init=10)
    kmeans_labels = kmeans_best.fit_predict(X_processed)
    
    print(f"‚úì KMeans –æ–±—É—á–µ–Ω —Å k={chosen_k}")
    print(f"  –†–∞–∑–º–µ—Ä—ã –∫–ª–∞—Å—Ç–µ—Ä–æ–≤: {np.bincount(kmeans_labels)}")
    
    # 6. DBSCAN –ö–õ–ê–°–¢–ï–†–ò–ó–ê–¶–ò–Ø
    print(f"\n{'='*60}")
    print(f"DBSCAN –ö–õ–ê–°–¢–ï–†–ò–ó–ê–¶–ò–Ø –î–ê–¢–ê–°–ï–¢–ê {dataset_num}")
    print(f"{'='*60}")
    
    print("üîç –ü–æ–¥–±–∏—Ä–∞–µ–º –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è DBSCAN...")
    
    # –ü–æ–¥–±–æ—Ä –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –¥–ª—è DBSCAN
    if dataset_num == 1:
        eps_values = np.linspace(0.1, 2.0, 20)
    elif dataset_num == 2:
        eps_values = np.linspace(0.1, 1.0, 20)  # –î–ª—è –Ω–µ–ª–∏–Ω–µ–π–Ω–æ–π —Å—Ç—Ä—É–∫—Ç—É—Ä—ã
    else:  # dataset_num == 3
        eps_values = np.linspace(0.05, 0.5, 20)  # –î–ª—è –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ —Ä–∞–∑–Ω–æ–π –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏
    
    min_samples_values = [3, 5, 10]
    dbscan_results = []
    
    for eps in eps_values:
        for min_samples in min_samples_values:
            dbscan = DBSCAN(eps=eps, min_samples=min_samples)
            labels = dbscan.fit_predict(X_processed)
            
            # –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞
            n_clusters = len(np.unique(labels[labels != -1]))
            noise_ratio = np.mean(labels == -1) * 100
            
            # –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º –º–µ—Ç—Ä–∏–∫–∏ —Ç–æ–ª—å–∫–æ –µ—Å–ª–∏ –µ—Å—Ç—å —Ö–æ—Ç—è –±—ã 2 –∫–ª–∞—Å—Ç–µ—Ä–∞
            if n_clusters > 1 and np.sum(labels != -1) > 1:
                mask = labels != -1
                silhouette = silhouette_score(X_processed[mask], labels[mask])
                db_score = davies_bouldin_score(X_processed[mask], labels[mask])
                ch_score = calinski_harabasz_score(X_processed[mask], labels[mask])
            else:
                silhouette = db_score = ch_score = np.nan
            
            dbscan_results.append({
                'eps': eps,
                'min_samples': min_samples,
                'n_clusters': n_clusters,
                'noise_ratio_%': noise_ratio,
                'silhouette': silhouette,
                'davies_bouldin': db_score,
                'calinski_harabasz': ch_score
            })
    
    # –°–æ–∑–¥–∞—ë–º DataFrame —Å —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º–∏
    dbscan_df = pd.DataFrame(dbscan_results)
    
    print(f"\nüìä –õ—É—á—à–∏–µ –∫–æ–Ω—Ñ–∏–≥—É—Ä–∞—Ü–∏–∏ DBSCAN –¥–ª—è {dataset_name}:")
    best_dbscan = dbscan_df.dropna().sort_values('silhouette', ascending=False).head()
    display(best_dbscan.round(4))
    
    # –í—ã–±–∏—Ä–∞–µ–º –ª—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è DBSCAN
    best_row = best_dbscan.iloc[0]
    dbscan_best = DBSCAN(eps=best_row['eps'], min_samples=int(best_row['min_samples']))
    dbscan_labels = dbscan_best.fit_predict(X_processed)
    
    print(f"\n‚úÖ –õ—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã DBSCAN –¥–ª—è {dataset_name}:")
    print(f"  eps = {best_row['eps']:.3f}, min_samples = {int(best_row['min_samples'])}")
    print(f"  –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤: {len(np.unique(dbscan_labels[dbscan_labels != -1]))}")
    print(f"  –î–æ–ª—è —à—É–º–∞: {np.mean(dbscan_labels == -1):.2%}")
    
    # 7. –í–ò–ó–£–ê–õ–ò–ó–ê–¶–ò–Ø –ö–õ–ê–°–¢–ï–†–û–í
    print(f"\n{'='*60}")
    print(f"–í–ò–ó–£–ê–õ–ò–ó–ê–¶–ò–Ø –ö–õ–ê–°–¢–ï–†–û–í –î–ê–¢–ê–°–ï–¢–ê {dataset_num}")
    print(f"{'='*60}")
    
    def plot_clusters_pca(X, labels, title, filename):
        """–í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ —Å –ø–æ–º–æ—â—å—é PCA"""
        # –ü—Ä–∏–º–µ–Ω—è–µ–º PCA –¥–ª—è —É–º–µ–Ω—å—à–µ–Ω–∏—è —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏ –¥–æ 2D
        pca = PCA(n_components=2, random_state=42)
        X_pca = pca.fit_transform(X)
        
        plt.figure(figsize=(12, 10))
        
        # –†–∞–∑–¥–µ–ª—è–µ–º —Ç–æ—á–∫–∏ –Ω–∞ –∫–ª–∞—Å—Ç–µ—Ä—ã –∏ —à—É–º (–µ—Å–ª–∏ –µ—Å—Ç—å)
        if -1 in labels:
            # –ï—Å—Ç—å —à—É–º
            noise_mask = labels == -1
            clusters_mask = labels != -1
            
            # –†–∏—Å—É–µ–º –∫–ª–∞—Å—Ç–µ—Ä—ã
            scatter_clusters = plt.scatter(
                X_pca[clusters_mask, 0], X_pca[clusters_mask, 1],
                c=labels[clusters_mask], cmap='tab20', alpha=0.8, s=50,
                edgecolors='w', linewidth=0.5
            )
            
            # –†–∏—Å—É–µ–º —à—É–º
            plt.scatter(
                X_pca[noise_mask, 0], X_pca[noise_mask, 1],
                c='black', marker='x', s=50, alpha=0.6,
                label=f'–®—É–º ({np.sum(noise_mask)} —Ç–æ—á–µ–∫)'
            )
            
            plt.legend()
        else:
            # –ù–µ—Ç —à—É–º–∞
            scatter_clusters = plt.scatter(
                X_pca[:, 0], X_pca[:, 1],
                c=labels, cmap='tab20', alpha=0.8, s=50,
                edgecolors='w', linewidth=0.5
            )
        
        plt.colorbar(scatter_clusters, label='–ú–µ—Ç–∫–∞ –∫–ª–∞—Å—Ç–µ—Ä–∞')
        plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.2%} –¥–∏—Å–ø–µ—Ä—Å–∏–∏)', fontsize=12)
        plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.2%} –¥–∏—Å–ø–µ—Ä—Å–∏–∏)', fontsize=12)
        plt.title(title, fontsize=14)
        plt.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.savefig(f'artifacts/figures/{filename}', dpi=300, bbox_inches='tight')
        plt.show()
        
        return pca
    
    print("üìä –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏...")
    
    # –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è KMeans
    pca_kmeans = plot_clusters_pca(
        X_processed, kmeans_labels,
        f'{dataset_name}: KMeans –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏—è (k={chosen_k})',
        f'{dataset_name}_pca_kmeans.png'
    )
    
    # –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è DBSCAN
    pca_dbscan = plot_clusters_pca(
        X_processed, dbscan_labels,
        f'{dataset_name}: DBSCAN –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏—è\n(eps={best_row["eps"]:.3f}, min_samples={int(best_row["min_samples"])})',
        f'{dataset_name}_pca_dbscan.png'
    )
    
    # 8. –ú–ï–¢–†–ò–ö–ò –ö–ê–ß–ï–°–¢–í–ê
    print(f"\n{'='*60}")
    print(f"–ú–ï–¢–†–ò–ö–ò –ö–ê–ß–ï–°–¢–í–ê –î–ê–¢–ê–°–ï–¢–ê {dataset_num}")
    print(f"{'='*60}")
    
    # –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Ä–∞—Å—á—ë—Ç–∞ –º–µ—Ç—Ä–∏–∫ —Å —É—á—ë—Ç–æ–º —à—É–º–∞
    def calculate_metrics_with_noise(X, labels, algorithm_name):
        """–†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ—Ç –º–µ—Ç—Ä–∏–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞ –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏ —Å —É—á—ë—Ç–æ–º —à—É–º–∞"""
        
        metrics = {
            'algorithm': algorithm_name,
            'n_clusters': len(np.unique(labels[labels != -1])) if -1 in labels else len(np.unique(labels)),
            'noise_ratio_%': np.mean(labels == -1) * 100 if -1 in labels else 0.0
        }
        
        # –ï—Å–ª–∏ –µ—Å—Ç—å —à—É–º –∏ –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ —Ç–æ—á–µ–∫ –¥–ª—è —Ä–∞—Å—á—ë—Ç–∞
        if -1 in labels and np.sum(labels != -1) > 1:
            mask = labels != -1
            if len(np.unique(labels[mask])) > 1:
                metrics['silhouette'] = silhouette_score(X[mask], labels[mask])
                metrics['davies_bouldin'] = davies_bouldin_score(X[mask], labels[mask])
                metrics['calinski_harabasz'] = calinski_harabasz_score(X[mask], labels[mask])
            else:
                metrics['silhouette'] = metrics['davies_bouldin'] = metrics['calinski_harabasz'] = np.nan
        elif -1 not in labels:
            # –ù–µ—Ç —à—É–º–∞
            metrics['silhouette'] = silhouette_score(X, labels)
            metrics['davies_bouldin'] = davies_bouldin_score(X, labels)
            metrics['calinski_harabasz'] = calinski_harabasz_score(X, labels)
        else:
            metrics['silhouette'] = metrics['davies_bouldin'] = metrics['calinski_harabasz'] = np.nan
        
        return metrics
    
    # –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º –º–µ—Ç—Ä–∏–∫–∏ –¥–ª—è –æ–±–æ–∏—Ö –∞–ª–≥–æ—Ä–∏—Ç–º–æ–≤
    metrics_kmeans = calculate_metrics_with_noise(
        X_processed, kmeans_labels, f'KMeans (k={chosen_k})'
    )
    
    metrics_dbscan = calculate_metrics_with_noise(
        X_processed, dbscan_labels, 
        f'DBSCAN (eps={best_row["eps"]:.3f}, min_samples={int(best_row["min_samples"])})'
    )
    
    # –°–æ–∑–¥–∞—ë–º DataFrame –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è
    comparison_df = pd.DataFrame([metrics_kmeans, metrics_dbscan])
    print(f"\nüìä –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫ –∫–∞—á–µ—Å—Ç–≤–∞ –¥–ª—è {dataset_name}:")
    display(comparison_df.round(4))
    
    # 9. –°–û–•–†–ê–ù–ï–ù–ò–ï –†–ï–ó–£–õ–¨–¢–ê–¢–û–í
    # –°–æ—Ö—Ä–∞–Ω—è–µ–º –º–µ—Ç–∫–∏ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ (–ª—É—á—à–µ–µ —Ä–µ—à–µ–Ω–∏–µ –ø–æ Silhouette)
    if metrics_kmeans['silhouette'] > metrics_dbscan['silhouette']:
        best_labels = kmeans_labels
        best_algorithm = 'KMeans'
    else:
        best_labels = dbscan_labels
        best_algorithm = 'DBSCAN'
    
    labels_df = pd.DataFrame({
        'sample_id': sample_ids if len(sample_ids) == len(best_labels) else range(len(best_labels)),
        'cluster_label': best_labels
    })
    
    labels_path = f'artifacts/labels/labels_hw07_ds{dataset_num}.csv'
    labels_df.to_csv(labels_path, index=False)
    print(f"‚úì –ú–µ—Ç–∫–∏ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã –≤: {labels_path}")
    print(f"  –õ—É—á—à–∏–π –∞–ª–≥–æ—Ä–∏—Ç–º: {best_algorithm}")
    
    # –í–æ–∑–≤—Ä–∞—â–∞–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã
    return {
        'dataset_name': dataset_name,
        'dataset_num': dataset_num,
        'shape': df.shape,
        'kmeans_metrics': metrics_kmeans,
        'dbscan_metrics': metrics_dbscan,
        'best_algorithm': best_algorithm,
        'best_labels': best_labels,
        'chosen_k': chosen_k,
        'dbscan_params': {'eps': best_row['eps'], 'min_samples': int(best_row['min_samples'])}
    }

# 10. –ê–ù–ê–õ–ò–ó –í–°–ï–• –î–ê–¢–ê–°–ï–¢–û–í
print("\n" + "="*80)
print("–ù–ê–ß–ê–õ–û –ê–ù–ê–õ–ò–ó–ê –í–°–ï–• –î–ê–¢–ê–°–ï–¢–û–í")
print("="*80)

# –°–ø–∏—Å–æ–∫ –¥–∞—Ç–∞—Å–µ—Ç–æ–≤ –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞
datasets = [
    {
        'name': 'S07-hw-dataset-01.csv',
        'path': 'data/S07-hw-dataset-01.csv',
        'num': 1,
        'description': '–ß–∏—Å–ª–æ–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –≤ —Ä–∞–∑–Ω—ã—Ö —à–∫–∞–ª–∞—Ö + —à—É–º–æ–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏'
    },
    {
        'name': 'S07-hw-dataset-02.csv',
        'path': 'data/S07-hw-dataset-02.csv',
        'num': 2,
        'description': '–ù–µ–ª–∏–Ω–µ–π–Ω–∞—è —Å—Ç—Ä—É–∫—Ç—É—Ä–∞ + –≤—ã–±—Ä–æ—Å—ã + –ª–∏—à–Ω–∏–π —à—É–º–æ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫'
    },
    {
        'name': 'S07-hw-dataset-03.csv',
        'path': 'data/S07-hw-dataset-03.csv',
        'num': 3,
        'description': '–ö–ª–∞—Å—Ç–µ—Ä—ã —Ä–∞–∑–Ω–æ–π –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ + —Ñ–æ–Ω–æ–≤—ã–π —à—É–º'
    }
]

# –ê–Ω–∞–ª–∏–∑–∏—Ä—É–µ–º –∫–∞–∂–¥—ã–π –¥–∞—Ç–∞—Å–µ—Ç
results = {}
for dataset_info in datasets:
    # –ü—Ä–æ–≤–µ—Ä—è–µ–º, —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –ª–∏ —Ñ–∞–π–ª
    if not os.path.exists(dataset_info['path']):
        print(f"\n‚ùå –§–∞–π–ª {dataset_info['path']} –Ω–µ –Ω–∞–π–¥–µ–Ω!")
        print(f"–ü–æ–∂–∞–ª—É–π—Å—Ç–∞, —Å–∫–∞—á–∞–π—Ç–µ {dataset_info['name']} –∏ –ø–æ–º–µ—Å—Ç–∏—Ç–µ –≤ –ø–∞–ø–∫—É data/")
        continue
    
    result = analyze_dataset(
        dataset_info['name'], 
        dataset_info['path'], 
        dataset_info['num']
    )
    
    if result:
        results[dataset_info['num']] = result
        results[dataset_info['num']]['description'] = dataset_info['description']

# 11. –ò–¢–û–ì–û–í–´–ô –ê–ù–ê–õ–ò–ó –ò –°–û–•–†–ê–ù–ï–ù–ò–ï –ê–†–¢–ï–§–ê–ö–¢–û–í
print("\n" + "="*80)
print("–ò–¢–û–ì–û–í–´–ô –ê–ù–ê–õ–ò–ó –ò –°–û–•–†–ê–ù–ï–ù–ò–ï –ê–†–¢–ï–§–ê–ö–¢–û–í")
print("="*80)

# –°–æ—Ö—Ä–∞–Ω—è–µ–º –º–µ—Ç—Ä–∏–∫–∏ –≤ JSON
metrics_summary = {}
best_configs = {}

for ds_num, result in results.items():
    dataset_key = f'dataset_{ds_num}'
    
    metrics_summary[dataset_key] = {
        'dataset_name': result['dataset_name'],
        'description': result['description'],
        'shape': list(result['shape']),
        'kmeans': result['kmeans_metrics'],
        'dbscan': result['dbscan_metrics'],
        'best_algorithm': result['best_algorithm']
    }
    
    best_configs[dataset_key] = {
        'best_algorithm': result['best_algorithm'],
        'kmeans': {
            'n_clusters': result['chosen_k'],
            'random_state': 42,
            'n_init': 10,
            'silhouette': result['kmeans_metrics']['silhouette']
        },
        'dbscan': {
            'eps': float(result['dbscan_params']['eps']),
            'min_samples': result['dbscan_params']['min_samples'],
            'silhouette': result['dbscan_metrics']['silhouette']
        }
    }

# –°–æ—Ö—Ä–∞–Ω—è–µ–º JSON —Ñ–∞–π–ª—ã
with open('artifacts/metrics_summary.json', 'w', encoding='utf-8') as f:
    json.dump(metrics_summary, f, indent=2, ensure_ascii=False)
print("‚úì –ú–µ—Ç—Ä–∏–∫–∏ —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã –≤: artifacts/metrics_summary.json")

with open('artifacts/best_configs.json', 'w', encoding='utf-8') as f:
    json.dump(best_configs, f, indent=2, ensure_ascii=False)
print("‚úì –ö–æ–Ω—Ñ–∏–≥—É—Ä–∞—Ü–∏–∏ —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã –≤: artifacts/best_configs.json")

# 12. –ü–†–û–í–ï–†–ö–ê –£–°–¢–û–ô–ß–ò–í–û–°–¢–ò (–¥–ª—è –æ–¥–Ω–æ–≥–æ –¥–∞—Ç–∞—Å–µ—Ç–∞)
print("\n" + "="*80)
print("–ü–†–û–í–ï–†–ö–ê –£–°–¢–û–ô–ß–ò–í–û–°–¢–ò (–¥–ª—è –¥–∞—Ç–∞—Å–µ—Ç–∞ 1)")
print("="*80)

if 1 in results:
    # –ò—Å–ø–æ–ª—å–∑—É–µ–º –æ–±—Ä–∞–±–æ—Ç–∞–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –∏–∑ –¥–∞—Ç–∞—Å–µ—Ç–∞ 1
    dataset_path = 'data/S07-hw-dataset-01.csv'
    df = pd.read_csv(dataset_path)
    
    if 'sample_id' in df.columns:
        X = df.drop('sample_id', axis=1)
    else:
        X = df.copy()
    
    # –ü—Ä–µ–ø—Ä–æ—Ü–µ—Å—Å–∏–Ω–≥
    numeric_features = X.select_dtypes(include=[np.number]).columns.tolist()
    numeric_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler())
    ])
    
    preprocessor = ColumnTransformer(
        transformers=[
            ('num', numeric_transformer, numeric_features)
        ],
        remainder='drop'
    )
    
    X_processed = preprocessor.fit_transform(X)
    
    # –ü—Ä–æ–≤–µ—Ä–∫–∞ —É—Å—Ç–æ–π—á–∏–≤–æ—Å—Ç–∏ KMeans
    n_runs = 5
    ari_scores = []
    chosen_k = results[1]['chosen_k']
    
    # –ü–µ—Ä–≤—ã–π –∑–∞–ø—É—Å–∫ –∫–∞–∫ —ç—Ç–∞–ª–æ–Ω
    kmeans_ref = KMeans(n_clusters=chosen_k, random_state=42, n_init=10)
    labels_ref = kmeans_ref.fit_predict(X_processed)
    
    different_seeds = [42, 123, 456, 789, 1000]
    
    print(f"üîç –ü—Ä–æ–≤–µ—Ä—è–µ–º —É—Å—Ç–æ–π—á–∏–≤–æ—Å—Ç—å KMeans —Å k={chosen_k}...")
    for seed in different_seeds:
        kmeans_temp = KMeans(n_clusters=chosen_k, random_state=seed, n_init=10)
        labels_temp = kmeans_temp.fit_predict(X_processed)
        ari = adjusted_rand_score(labels_ref, labels_temp)
        ari_scores.append(ari)
        print(f"  random_state={seed:4d}: ARI = {ari:.4f}")
    
    print(f"\nüìä –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ —É—Å—Ç–æ–π—á–∏–≤–æ—Å—Ç–∏:")
    print(f"  –°—Ä–µ–¥–Ω–∏–π ARI: {np.mean(ari_scores):.4f}")
    print(f"  –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ: {np.std(ari_scores):.4f}")
    print(f"  –ú–∏–Ω–∏–º–∞–ª—å–Ω—ã–π ARI: {np.min(ari_scores):.4f}")
    print(f"  –ú–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π ARI: {np.max(ari_scores):.4f}")
    
    # –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —É—Å—Ç–æ–π—á–∏–≤–æ—Å—Ç–∏
    plt.figure(figsize=(10, 6))
    plt.bar(range(len(ari_scores)), ari_scores, color='lightblue', edgecolor='black', alpha=0.8)
    plt.axhline(y=np.mean(ari_scores), color='r', linestyle='--', 
               label=f'–°—Ä–µ–¥–Ω–µ–µ: {np.mean(ari_scores):.4f}')
    plt.xlabel('–ù–æ–º–µ—Ä –∑–∞–ø—É—Å–∫–∞', fontsize=12)
    plt.ylabel('Adjusted Rand Index (ARI)', fontsize=12)
    plt.title('–£—Å—Ç–æ–π—á–∏–≤–æ—Å—Ç—å KMeans: —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ä–∞–∑–±–∏–µ–Ω–∏–π (–î–∞—Ç–∞—Å–µ—Ç 1)', fontsize=14)
    plt.xticks(range(len(ari_scores)), [f'–ó–∞–ø—É—Å–∫ {i+1}' for i in range(len(ari_scores))])
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig('artifacts/figures/kmeans_stability.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # –°–æ—Ö—Ä–∞–Ω—è–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —É—Å—Ç–æ–π—á–∏–≤–æ—Å—Ç–∏
    stability_results = {
        'dataset': 'S07-hw-dataset-01.csv',
        'n_clusters': int(chosen_k),
        'n_runs': n_runs,
        'ari_scores': [float(score) for score in ari_scores],
        'mean_ari': float(np.mean(ari_scores)),
        'std_ari': float(np.std(ari_scores))
    }
    
    metrics_summary['stability_test'] = stability_results
    with open('artifacts/metrics_summary.json', 'w', encoding='utf-8') as f:
        json.dump(metrics_summary, f, indent=2, ensure_ascii=False)
    
    print("‚úì –†–µ–∑—É–ª—å—Ç–∞—Ç—ã —É—Å—Ç–æ–π—á–∏–≤–æ—Å—Ç–∏ —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã –≤ metrics_summary.json")

# 13. –§–ò–ù–ê–õ–¨–ù–´–ô –û–¢–ß–Å–¢
print("\n" + "="*80)
print("–§–ò–ù–ê–õ–¨–ù–´–ô –û–¢–ß–Å–¢")
print("="*80)

print(f"\nüìã –ê–ù–ê–õ–ò–ó –í–´–ü–û–õ–ù–ï–ù –î–õ–Ø {len(results)} –î–ê–¢–ê–°–ï–¢–û–í:")
for ds_num, result in results.items():
    print(f"\n–î–∞—Ç–∞—Å–µ—Ç {ds_num}: {result['dataset_name']}")
    print(f"  –†–∞–∑–º–µ—Ä: {result['shape'][0]} —Å—Ç—Ä–æ–∫, {result['shape'][1]} —Å—Ç–æ–ª–±—Ü–æ–≤")
    print(f"  –û–ø–∏—Å–∞–Ω–∏–µ: {result['description']}")
    print(f"  –õ—É—á—à–∏–π –∞–ª–≥–æ—Ä–∏—Ç–º: {result['best_algorithm']}")
    print(f"  KMeans k: {result['chosen_k']}")
    print(f"  DBSCAN params: eps={result['dbscan_params']['eps']:.3f}, min_samples={result['dbscan_params']['min_samples']}")

print("\nüìà –°–û–ó–î–ê–ù–ù–´–ï –ì–†–ê–§–ò–ö–ò:")
figures = os.listdir('artifacts/figures')
for fig in sorted(figures):
    print(f"  ‚úì {fig}")

print("\nüè∑Ô∏è –°–û–•–†–ê–ù–Å–ù–ù–´–ï –ú–ï–¢–ö–ò –ö–õ–ê–°–¢–ï–†–û–í:")
labels_files = os.listdir('artifacts/labels')
for label_file in sorted(labels_files):
    print(f"  ‚úì {label_file}")

print("\n‚úÖ –í–´–ü–û–õ–ù–ï–ù–ò–ï –ó–ê–í–ï–†–®–ï–ù–û!")
print("–í—Å–µ –∞—Ä—Ç–µ—Ñ–∞–∫—Ç—ã —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã –≤ –ø–∞–ø–∫–µ artifacts/")
print(f"–ü—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞–Ω–æ –¥–∞—Ç–∞—Å–µ—Ç–æ–≤: {len(results)} –∏–∑ {len(datasets)}")
```
